# 🛰️ Afet-Rota-AI: Kapsamlı Rota Optimizasyon Simulasyonu

Bu notebook, **Beşiktaş, İstanbul** bölgesinde bir afet senaryosunu ve **Afet-Rota-AI**'nın bu duruma nasıl tepki verdiğini simüle eder.

```python
import os
import sys
sys.path.append('..')

import osmnx as ox
import matplotlib.pyplot as plt
from src.routing.graph_utils import GraphLoader
from src.routing.main_navigator import RouteEngine
from src.vision.damage_assessment import DamageScanner
```

## 1. Veri Hazırlığı (OSM Graph Yükleme)
Beşiktaş bölgesinin yol ağını indiriyoruz.

```python
loader = GraphLoader()
G = loader.load_from_address("Besiktas, Istanbul", dist=1500)
engine = RouteEngine(G)

print(f"[+] Graph Yüklendi: {len(G.nodes)} düğüm, {len(G.edges)} kenar.")
```

## 2. Afet Öncesi: Standart Rota
Hiçbir engel yokken en kısa yolu belirliyoruz.

```python
start_coords = (41.042, 29.008) # Örn: Barbaros Bulvarı başı
end_coords = (41.048, 29.002)   # Örn: Dikilitaş

print("[*] Standart Rota Hesaplanıyor...")
standard_route = engine.find_safe_route(start_coords, end_coords)
print(f"[+] Standart Rota Bulundu: {len(standard_route)} düğüm.")
```

## 3. Afet Anı: Uydu Verisi ile Engel Tespiti
Sistem uydu görüntüsünü analiz eder ve bir enkaz tespit eder.

```python
scanner = DamageScanner()
detections = scanner.analyze_image(None) # Simüle edilmiş analiz

blocked_points = [d['coords'] for d in detections]
engine.apply_disaster_risk(blocked_points)

print(f"[!] {len(blocked_points)} adet engel sisteme işlendi.")
```

## 4. Afet Sonrası: Güvenli Rota
Engelleri dikkate alan yeni rotayı hesaplıyoruz.

```python
print("[*] Güvenli Rota Hesaplanıyor...")
safe_route = engine.find_safe_route(start_coords, end_coords)
print(f"[+] Güvenli Rota Bulundu: {len(safe_route)} düğüm.")
if safe_route != standard_route:
    print("✅ Başarılı: Sistem engelden kaçınan alternatif bir rota oluşturdu.")
```
